### 栈

```python
class StackUnderflow(ValueError):
    pass

class SStack:
    '''This is a stack which is complement by list'''
    def __init__(self):
        self.items = []
    
    def is_empty(self):
        return len(self.item) == 0
    
    def push(self, item):
        self.items.append(item)
        
    def pop(self):
        self.items.pop()
        
    def top(self):
        if len(self.items) == 0:
            raise StackUnderflow("This stack's length is 0")
        return self.items[-1]
    
class LStack:
    '''This is a stack which is complement by list'''
    def __init__(self):
        self._top = None
        
    def is_empty(self):
        return self._top is None
    
    def push(self, item):
        pass
    
    def pop(self):
        pass
    
    def top(self):
        pass
```
